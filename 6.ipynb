```python
import numpy as np
import matplotlib.pyplot as plt

modelos = ["Árbol", "Bosque", "k-NN", "Weighted k-NN"]
acc_val_mean = [0.94, 0.98, 0.97, 0.97]
acc_val_std  = [0.02, 0.011, 0.01, 0.01]
acc_test     = [0.96, 0.98, 0.92, 0.92]

x = np.arange(len(modelos))
width = 0.35  # ancho de las barras

plt.figure(figsize=(8,5))
plt.bar(x - width/2, acc_val_mean, width, yerr=acc_val_std, capsize=4, label="Validación", color="#B0B969")
plt.bar(x + width/2, acc_test, width, label="Test", color="#003732")

plt.ylabel("Accuracy")
#plt.title("Comparación de modelos")
plt.xticks(x, modelos)
plt.ylim(0.9, 1.0)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
#plt.savefig('Val_vs_Test.png', dpi=1000)
plt.show()
```
